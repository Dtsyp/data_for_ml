# EDA — Exploratory Data Analysis

This notebook demonstrates the DataCollectionAgent and EDA capabilities.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

## 1. Load Data

```python
df = pd.read_parquet('../data/raw/combined.parquet')
print(f'Dataset: {len(df)} rows, {len(df.columns)} columns')
print(f'Columns: {list(df.columns)}')
df.head()
```

## 2. Class Distribution

```python
label_counts = df['label'].value_counts()
print(label_counts)

fig, ax = plt.subplots(figsize=(10, 6))
label_counts.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_title('Class Distribution')
ax.set_xlabel('Class')
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3. Text Length Distribution

```python
text_lengths = df['text'].str.len()
print(f'Mean: {text_lengths.mean():.0f}, Std: {text_lengths.std():.1f}')
print(f'Min: {text_lengths.min()}, Max: {text_lengths.max()}')

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(text_lengths, bins=30, color='steelblue', edgecolor='black')
ax.set_title('Text Length Distribution')
ax.set_xlabel('Characters')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4. Top 20 Words

```python
all_words = ' '.join(df['text'].str.lower()).split()
stop_words = {'the','a','an','is','it','to','of','and','in','for','on','with','that','this','was','are','be'}
filtered = [w for w in all_words if w not in stop_words]
top_20 = Counter(filtered).most_common(20)

words, counts = zip(*top_20)
fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(list(reversed(words)), list(reversed(counts)), color='steelblue', edgecolor='black')
ax.set_title('Top 20 Words')
ax.set_xlabel('Count')
plt.tight_layout()
plt.show()
```
